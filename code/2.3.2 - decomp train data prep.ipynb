```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf

data = pd.read_csv("../data/decomp/kusok_1.csv")  # Загрузить файл
season_periods = {}
data.drop(columns=['Unnamed: 0'], inplace=True)
for col in data.columns:
    if col == 'full_date':
        continue
    series = data[col]  # Выбрать временной ряд
    lags = 500  # Достаточно, чтобы захватить возможный сезон
    
    acf_values = acf(series, nlags=lags)
    season_length = acf_values[100:].argmax() + 100
    print(season_length)
    season_periods[col] = season_length
```

```python
import numpy as np
import pandas as pd

# Определяем среднюю длину сезона
season_length = 360

# Датасет 1: Центрирование на середине данных
centered_data_1 = {}

for col in data.columns:
    if col == 'full_date':
        continue

    series = data[col].dropna()
    N = len(series)

    # Центрируем на середине
    start_idx = max(0, (N - season_length) // 2)
    end_idx = start_idx + season_length

    centered_data_1[col] = series.iloc[start_idx:end_idx].reset_index(drop=True)

df_centered_1 = pd.DataFrame(centered_data_1)
df_centered_1.to_csv("../data/decomp/dataset_centered_middle.csv", index=False)
```

```python
# Датасет 2: Индивидуальные сезоны
centered_data_2 = {}

for col in data.columns:
    if col == 'full_date':
        continue

    series = data[col].dropna()
    season_length = season_periods.get(col, 360)  # Если нет в словаре, используем 360

    N = len(series)

    # Центрируем на середине с индивидуальным сезоном
    start_idx = max(0, (N - season_length) // 2)
    end_idx = start_idx + season_length

    centered_data_2[col] = series.iloc[start_idx:end_idx].reset_index(drop=True)

df_centered_2 = pd.DataFrame(centered_data_2)
df_centered_2.to_csv("../data/decomp/dataset_centered_individual.csv", index=False)
```
